## Задание 3

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

```python
data = df.drop(columns=['id','zipcode','date']).reset_index()
corr_matrix = data.corr()
corr_matrix = np.round(corr_matrix, 1)
#corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(15, 15))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

Из данной матрицы мы можем увидеть что наибольшее влияние на стоимость дома оказывают такие параметры как:  
sqft_above - площать дома выше уровня земли  
grade - уровень конструкции и дизайна  
sqft_living - количество жилой площади  
bathrooms - количество ванных комнат  
sqft_living_15 но влияние этого параметра вызывает вопросы
